```python
# merge the datasets using 'id'
```

```python
merged_data = pd.merge(client_data,price_data, on='id')
```

```python
merged_data.sample(3)
```

```python
# Convert date into datetime from object type
```

```python
merged_data['date_activ'] = pd.to_datetime(merged_data['date_activ'])
merged_data['date_end'] = pd.to_datetime(merged_data['date_activ'])
merged_data['date_modif_prod'] = pd.to_datetime(merged_data['date_activ'])
merged_data['date_renewal'] = pd.to_datetime(merged_data['date_activ'])
merged_data['price_date'] = pd.to_datetime(merged_data['date_activ'])
```

```python
merged_data.info()
```

```python
# Data Visualization
```

```python
# 1. Power and Gas Consumption in the past 12 months
```

```python
merged_data.sort_values(by='price_date', inplace=True)
power_consumption = merged_data[['price_date', 'cons_12m']]
gas_consumption = merged_data[['price_date', 'cons_gas_12m']]

plt.figure(figsize=(10, 6))
plt.plot(power_consumption['price_date'], power_consumption['cons_12m'], label='Power Consumption', color='blue')
plt.plot(gas_consumption['price_date'], gas_consumption['cons_gas_12m'], label='Gas Consumption', color='orange')

plt.xlabel('Date')
plt.ylabel('Consumption')
plt.title('Power and Gas Consumption in the Last 12 Months')
plt.legend()

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# 2. Sales Channel
```

```python
# count the occurences of sales channels
sales_channel_count = client_data['channel_sales'].value_counts()
plt.figure(figsize=(8,6))
sns.barplot(x=sales_channel_count.index, y=sales_channel_count.values, palette='viridis')

plt.xlabel('Sales Channel')
plt.ylabel('Number of clients')
plt.title('Distribution of sales channels')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 3. Distribution of Gas Clients 
```

```python
# Count the occurrences of 'has_gas' values
gas_client_counts = client_data['has_gas'].value_counts()
plt.figure(figsize=(6, 6))
sns.barplot(x=gas_client_counts.index, y=gas_client_counts.values, palette='magma')

#labels and title
plt.xlabel('Has Gas Client')
plt.ylabel('Number of Clients')
plt.title('Distribution of Gas Clients')
plt.tight_layout()
plt.show()
```

```python
# 4. Churn Rate
```

```python
churn_counts = merged_data['churn'].value_counts()
plt.figure(figsize=(6, 6))  
plt.bar(['Churned', 'Retained'], churn_counts.values, color=['red', 'green'])
plt.xlabel('Churn Status')
plt.ylabel('Number of Customers')
plt.title('Churn Visualization')
for i, count in enumerate(churn_counts):
    plt.text(i, count, str(count), ha='center', va='bottom', fontsize=12, fontweight='bold')
    plt.tight_layout() 
plt.show()
```

```python
# 5. Forcasted Vs. Actual electricity cunsumption in last 12 months
```

```python
plt.figure(figsize=(12, 6)) 
plt.plot(merged_data['date_activ'], merged_data['cons_12m'], marker='o', color='b', label='Actual Consumption')
plt.plot(merged_data['date_activ'], merged_data['forecast_cons_12m'], marker='o', color='r', label='Forecasted Consumption')
plt.xlabel('Date')
plt.ylabel('Electricity Consumption')
plt.title('Actual vs. Forecasted Electricity Consumption ')
plt.legend()
plt.grid(True) 
plt.tight_layout() 
plt.xticks(rotation=45) 
plt.show()
```

```python
# 6. Margin
```

```python
plt.figure(figsize=(8, 6))
plt.bar(['Gross Margin', 'Net Margin'],
        [merged_data['margin_gross_pow_ele'].mean(), merged_data['margin_net_pow_ele'].mean()],
        color=['blue', 'green'])
plt.xlabel('Margin Type')
plt.ylabel('Average Margin')
plt.title('Comparison of Gross and Net Margins')
plt.tight_layout()  
plt.show()
```

```python
# 7. Distribution of margin to identify outliers
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=merged_data[['margin_gross_pow_ele', 'margin_net_pow_ele']], palette='pastel')
plt.ylabel('Margin')
plt.title('Distribution of Gross and Net Margins')
plt.show()
```

```python
# 8. Relationship between Margin and Electricity Consumption
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(merged_data['cons_12m'], merged_data['margin_gross_pow_ele'], color='blue', label='Gross Margin')
plt.scatter(merged_data['cons_12m'], merged_data['margin_net_pow_ele'], color='green', label='Net Margin')
plt.xlabel('Electricity Consumption')
plt.ylabel('Margin')
plt.title('Margins vs. Electricity Consumption')
plt.legend()
plt.show()
```
